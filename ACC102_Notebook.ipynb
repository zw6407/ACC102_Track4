```python
# ACC102 Mini Assignment
# Track 4: Interactive Data Analysis Tool
# Student ID: 2470452
# Student Name: Zhuowei.Li

# 1. Problem Definition
# The goal is to build an interactive tool for stock price analysis.
# Target users: finance students, individual investors.

# 2. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta

# 3. Generate simulated MSFT stock data (100% local, no external files, no internet)
# Create date range from 2023-01-01 to 2026-01-01
start_date = datetime(2023, 1, 1)
end_date = datetime(2026, 1, 1)
date_range = [start_date + timedelta(days=x) for x in range((end_date - start_date).days)]

# Generate realistic stock price trend (simulate MSFT's historical growth)
np.random.seed(42)  # Fix random seed for reproducibility
base_price = 240  # Starting price (MSFT 2023 opening price)
daily_returns = np.random.normal(0.0005, 0.015, len(date_range))  # Daily return with 0.05% drift
price_series = base_price * (1 + daily_returns).cumprod()

# Create DataFrame with standard stock data columns
data = pd.DataFrame({
    'Date': date_range,
    'Open': price_series * np.random.uniform(0.998, 1.002, len(date_range)),
    'High': price_series * np.random.uniform(1.001, 1.005, len(date_range)),
    'Low': price_series * np.random.uniform(0.995, 0.999, len(date_range)),
    'Close': price_series,
    'Adj Close': price_series,
    'Volume': np.random.randint(10000000, 50000000, len(date_range))
})

# 4. Data Cleaning
data = data.dropna()

# 5. Calculate Key Metrics
max_price = data['Close'].max()
min_price = data['Close'].min()
avg_price = data['Close'].mean()

print("===== Key Metrics =====")
print(f"Highest Close Price: ${max_price:.2f}")
print(f"Lowest Close Price: ${min_price:.2f}")
print(f"Average Close Price: ${avg_price:.2f}")

# 6. Visualization
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'], label='MSFT Close Price', color='blue')
plt.title('Simulated MSFT Stock Price Trend (2023-2026)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
